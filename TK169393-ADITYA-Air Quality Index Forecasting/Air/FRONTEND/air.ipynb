## LABEL ENCODING

Label encoding is a technique in machine learning used to convert categorical data into numerical form. Each unique category is assigned a unique integer label. It's suitable for ordinal categorical data where the categories have a natural order. However, it may introduce unintended ordinal relationships where none exist. For instance, converting categories like "low," "medium," and "high" to 0, 1, and 2 implies an inherent order. This method is efficient for algorithms that can handle numerical inputs but may not be ideal for algorithms sensitive to magnitude, as they might misinterpret the encoded values as having meaningful differences.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['City'] = le.fit_transform(df['City'])
df['PM2.5'] = le.fit_transform(df['PM2.5'])
df['PM10'] = le.fit_transform(df['PM10'])
df['NO'] = le.fit_transform(df['NO'])
df['NO2'] = le.fit_transform(df['NO2'])
df['NOx'] = le.fit_transform(df['NOx'])
df['NH3'] = le.fit_transform(df['NH3'])
df['CO'] = le.fit_transform(df['CO'])
df['SO2'] = le.fit_transform(df['SO2'])
df['O3'] = le.fit_transform(df['O3'])
df['Benzene'] = le.fit_transform(df['Benzene'])
df['Toluene'] = le.fit_transform(df['Toluene'])
df['Xylene'] = le.fit_transform(df['Xylene'])
df['AQI'] = le.fit_transform(df['AQI'])
df['AQI_Bucket'] = le.fit_transform(df['AQI_Bucket'])
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Example: Generate a synthetic time series data
data = np.sin(np.linspace(0, 100, num=1000))  # Replace this with your data
df = pd.DataFrame(data, columns=['values'])

# Split the data into training and testing sets
train_size = int(len(df) * 0.8)
train_data, test_data = df[:train_size], df[train_size:]

# Define the sequence length
seq_length = 10  # Adjust this based on your data and problem

# Prepare sequences for LSTM
def prepare_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequence = data[i:i+seq_length]
        target = data[i+seq_length]
        sequences.append((sequence, target))
    return np.array([s[0] for s in sequences]), np.array([s[1] for s in sequences])

X_train, y_train = prepare_sequences(train_data.values, seq_length)
X_test, y_test = prepare_sequences(test_data.values, seq_length)

# Reshape input data for LSTM
X_train = X_train.reshape(-1, seq_length, 1)
X_test = X_test.reshape(-1, seq_length, 1)

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(seq_length, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Evaluate the model on test data
test_loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")

# Make predictions
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Compute Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae}")

# Compute Mean Squared Error (MSE)
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")

# Compute Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```
