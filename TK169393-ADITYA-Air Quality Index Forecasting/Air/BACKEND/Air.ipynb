```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"city_day.csv")
df.head()
```

```python
### Checking the information about dataset [ data types , null values]
df.info()
```

```python
### Checking the null values
df.isna().sum()
```

```python
### filiing the null values
df['PM2.5'].fillna(df['PM2.5'].mean(), inplace=True)
df['PM10'].fillna(df['PM10'].mean(), inplace=True)
df['NO'].fillna(df['NO'].mean(), inplace=True)
df['NO2'].fillna(df['NO2'].mean(), inplace=True)
df['NOx'].fillna(df['NOx'].mean(), inplace=True)
df['NH3'].fillna(df['NH3'].mean(), inplace=True)
df['CO'].fillna(df['CO'].mean(), inplace=True)
df['SO2'].fillna(df['SO2'].mean(), inplace=True)
df['O3'].fillna(df['O3'].mean(), inplace=True)
df['Benzene'].fillna(df['Benzene'].mean(), inplace=True)
df['Toluene'].fillna(df['Toluene'].mean(), inplace=True)
df['Xylene'].fillna(df['Xylene'].mean(), inplace=True)
df['AQI'].fillna(df['AQI'].mean(), inplace=True)
# df['AQI_Bucket'].fillna(df['AQI_Bucket'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df["AQI_Bucket"].value_counts().to_dict()
```

```python
# Proper Sequence nut we conver 6 into 3 by comining 2 

# Very Poor     val = 0
# Poor          val = 0
# Satisfactory  val = 1
# Moderate      val = 1
# Good          val = 2
# Severe        val = 2
```

```python
df["AQI_Bucket"].replace({'Moderate': 1, 'Satisfactory': 1, 'Poor': 0,
                           'Very Poor': 0,  'Good': 2, 'Severe': 2}, inplace=True)
```

```python
df['AQI_Bucket'].fillna(df['AQI_Bucket'].ffill(), inplace=True)
df['AQI_Bucket'].fillna(df['AQI_Bucket'].bfill(), inplace=True)
```

```python
### Cheking the null value present or not after filling the null values
df.isna().sum()
```

```python
### cheking datatype and INFORMATION of each columns
df.info()
```

```python
df['City'].value_counts()
```

```python
### we drop the columns which is not that much importance 

df.drop(["City", "Date"], axis=1,inplace=True)
```

```python
### Splitting the Data
x = df.drop("AQI_Bucket", axis=1)
y = df["AQI_Bucket"]
```

```python
df['AQI_Bucket'].value_counts()
```

```python
### data is imbalance so we are balance the data using the SMOTE technic
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to your features (X) and target labels (y)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
### Splitting the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x_resampled, y_resampled, test_size = 0.3, stratify = y_resampled, random_state = 42 )
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```
